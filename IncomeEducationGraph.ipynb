```python
import numpy as np # import numpy package which allows us to conduct mathematical operations for the defined quantities.
import matplotlib.pyplot as plt# import matplot in order to be able to plot data points on a graph
```

```python
Income_arr = np.array([36700.,46400.,50200.,54200.,75100.,89200.]) # create an array that represents the data points for men's income
Education_arr = np.arange(len(Income_arr)) #an array for the education levels of the men
plt.plot(Education_arr,Income_arr, "o-") # plot the income and education array
plt.xlabel("Education Level") # x title
plt.ylabel("Income [$]") # y title
plt.show() # show the graph
```

```python
# assign labels that represent the education levels for the men
Education_levels = np.array([
"Less than high school completion",
"High school completion",
"Some college, no degree",
"Associate’s degree",
"Bachelor’s degree",
"Master’s or higher degree"
])
```

```python
# plot the graph with the labels for the education levels
plt.plot(Education_arr,Income_arr, "o-")
plt.xlabel("Education Level")
plt.ylabel("Income [$]")
plt.xticks(Education_arr, Education_levels, rotation=45, ha="right")
plt.show()
```

```python
z_linear = np.polyfit(Education_arr,Income_arr,1) # Use polyfit to fit the data with a linear relathioship
# "income = z_linear[0] * level + z_linear[1]", where z_linear[0] represents the slope and z_linear[1] represents the y-intercept
print("Slope = ", z_linear[0], " in dollar per level")
print("Y-intercept = ", z_linear[1], " in dollar at level 0, i.e., Less than high school completion")
```

```python
z_quadratic = np.polyfit(Education_arr,Income_arr,2)# Use polyfit to fit the data with a quadratic relathioship
print(z_quadratic)
print("That is: Income = {:.1f} * level**2 + {:.1f} * level + {:.1f}".format(z_quadratic[0], z_quadratic[1], z_quadratic[2]))
# Note the above line on the usage of {} for printing variables, and :.1f to control the printed number of siginifiant digits
```

```python
# Plot the data and the two fits: linear and quadratic
# Rename some variables to be consistent with the above definition
O = Income_arr
O_err = np.array([3110, 1070, 2000, 4130, 4110, 8910]) # extract from the same figure with Confidence interval toggled on, and remove the first entry
C_linear = z_linear[0] * Education_arr + z_linear[1]
C_quadratic = z_quadratic[0] * Education_arr**2 + z_quadratic[1] * Education_arr + z_quadratic[2]

plt.errorbar(Education_arr, O, yerr=O_err, marker="o", linestyle = "None", color = "black", label = "Data") # plot the error bars
plt.plot(Education_arr,Income_arr, color = "green", label = "Income vs education level") # plot the income and education array
plt.plot(Education_arr,C_linear, color = "red", label = "Linear fit") # print the linear fit for the original data
plt.plot(Education_arr,C_quadratic, color = "blue", label = "Quadratic fit") # print the quadratic fit for the original data
plt.xticks(Education_arr, Education_levels, rotation=45, ha="right")
plt.legend()
plt.xlabel("Education Level")
plt.ylabel("Income [$]")
plt.show()
```
