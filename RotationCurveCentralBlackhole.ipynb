```python
import astropy.units as u # import a python package to call certain units\
import astropy.constants as ac # import python package to call the value of certain constants
import numpy as np # import numpy package which allows us to conduct mathematical operations for the defined quantities.
```

How to define a function?

```python
def calculatingOrbitalVelocity(M, R):
  """
  This is a function that calculates the orbital velocity
  Input:
  M, the mass of the central object
  R, the orbital radius of the satellite
  Output:
  orbitalVelocity, the orbital velocity of the satellite
  """
  orbitalVelocity = np.sqrt(ac.G * M / R) # The formula for calculating orbital velocity
  return orbitalVelocity # return the orbital velocity to the function that called this one
```

```python
#Below we define input M and R
M = 1 * u.earthMass # get the mass of earth
R = 1 * u.earthRad # get the radius of earth

#Calculating orbital velocity of an earth satellite
v_earthSatellite = calculatingOrbitalVelocity(M, R) # using the predefined function
print(v_earthSatellite.to(u.km/u.s)) # print the orbital velocity in km/s
```

We are going to introduce the concept of an array

```python
# Mercury, Venus, Earth, Mars, Jupiter, Saturn, Uranus, Neptune
R_planet_arr = np.array([0.4, 0.7, 1.0, 1.5, 5.2, 9.6, 19.2, 30.1]) * u.au# Orbital radius of the eight solar system planets from inside out
```

```python
# Now we calculate the orbital velocities for eight planets
M_sun = 1 * u.solMass# mass of the sun
v_planet_arr = calculatingOrbitalVelocity(M_sun, R_planet_arr) # calculate orbital velocity of those planets around the sun
print(v_planet_arr.to(u.km/u.s)) # print out the orbital velocities of each planet around the sun
```

```python
# prompt: Can you help me with plotting v_planet_arr and R_planet_arr?

import matplotlib.pyplot as plt # importing plot package

plt.plot(R_planet_arr, v_planet_arr.to(u.km/u.s)) # plotting velocity vs orbital radius
plt.xlabel("Orbital Radius (AU)".format(R_planet_arr.unit)) # Add x axis label
plt.ylabel("Orbital Velocity (km/s)".format(v_planet_arr.unit)) # Add y axis label
plt.title("Orbital Velocity vs Orbital Radius for Planets") # Add title
plt.show() # displaying the graph
```

```python
#Radius for Milky Way: Array from 1 to 30 kiloparsecs
distance_from_blackhole = np.arange(1,30) * u.kiloparsec
#Mass of Sagittarius A: 4.2 million solar masses
mass_of_SagittariusA = 4297000 * u.solMass
rotational_curve_arr = calculatingOrbitalVelocity(mass_of_SagittariusA, distance_from_blackhole)# calculate the orbital velocity at different positions in the galaxy
print(rotational_curve_arr.to(u.km/u.s))# print out all the rotational velocities
```

```python
# prompt: Can you help me with plotting distance_from_blackhole and rotational_curve?

import matplotlib.pyplot as plt # importing plot package

plt.plot(distance_from_blackhole, rotational_curve_arr.to(u.km/u.s)) # plotting velocity vs orbital radius
plt.xlabel("Distance from Blackhole (kpc)") # Add x axis label
plt.ylabel("Rotational Velocity (km/s)") # Add y axis label
plt.title("Rotational Velocity vs Distance from Blackhole") # Add title
plt.show() #display the graph
```
